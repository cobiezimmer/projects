```python
import lyricsgenius as lg
genius = lg.Genius(#genius_secret,
                   skip_non_songs=True, 
                   remove_section_headers=True,
                   timeout=3600)
```

```python
songs = genius.search_artist('Taylor Swift', max_songs=500, sort='popularity').songs
```

```python
import pickle 

# to save file
with open('songs_swift.pkl', 'wb') as f:
    pickle.dump(songs, f)

# to open file
#with open('songs_swift.pkl', 'rb') as f:
#    songs = pickle.load(f)
```

```python
import re

rx = re.compile('\W+')

def decode(coded_abbr):
    c=0
    for song in songs:
        lyrics = song.lyrics
        lyrics = lyrics.replace("'", '')
        lyrics = rx.sub(' ', lyrics).strip()
        toks = lyrics.split()
        abbr = ''.join([x[0].lower() for x in toks])
        if coded_abbr.lower() in abbr:
            print(song.title)
            print('index is:', c)
            i = abbr.index(coded_abbr.lower())
            print(' '.join(toks[i:i+len(coded_abbr)]))
            print()
        c+=1
```
